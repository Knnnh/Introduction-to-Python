```python
#Solution 1
def list_to_dict1(my_list):
    my_dict={}
    for element in my_list:
        my_dict[element] = my_list.count(element)
    return my_dict
```

```python
list_to_dict1(list_test)
```

```python
#Solution 2

def list_to_dict2(my_list):
    set_list = set(my_list)
    new_dict ={}
    for element in set_list:
        count =0
        for i in my_list:
            if element ==i:
                count = count+1
        new_dict[element] = count 
    return new_dict
```

```python
list_to_dict2(list_test)
```

```python
def clean_string(string):
    prohib_carac = "0123456789@#&\"'(!){}[]-_=+:/%;.,?<>"
    lower_clean_string = ""
    for char in string:
        if char not in prohib_carac:
            lower_clean_string = lower_clean_string + char.lower()
    lower_clean_string_list = lower_clean_string.split()
    return lower_clean_string_list
```
