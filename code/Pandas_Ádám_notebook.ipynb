# Merging dataframes
<br>Documentation is here: https://pandas.pydata.org/docs/user_guide/merging.html

```python
samsung_reviews = data_phones[data_phones["Brand Name"] == "Samsung"][["Brand Name","Reviews"]].reset_index(drop=True)
nokia_reviews = data_phones[data_phones["Brand Name"] == "Nokia"][["Brand Name","Reviews"]].reset_index(drop=True)
```

```python
pd.concat([samsung_reviews,nokia_reviews], axis = 1)
```

```python
#Create function to create dataframe with rating distribution, sorted from lowest to highest

def create_rate_dist_df(brand_name):
    review_rate = data_phones[data_phones["Brand Name"] == brand_name]["Rating"].value_counts()
    review_rate = pd.DataFrame(review_rate).reset_index()
    review_rate["Type"] = [brand_name]*len(review_rate)
    review_rate = review_rate.rename(columns = {"index":"Rating", "Rating":"Number of rating"})
    review_rate = review_rate.sort_values(by="Rating")
    review_rate["% of total rates"] = review_rate["Number of rating"]/review_rate["Number of rating"].sum()
    review_rate = review_rate.reset_index(drop=True)
    
    return review_rate
```

```python
#Generate the function for multiple brands
types = ["Nokia", "Samsung"]
dataframe_df = {}

for i in types:
    dataframe_df[i] = create_rate_dist_df(i)
```

```python
pd.concat([dataframe_df["Nokia"],dataframe_df["Samsung"]], axis = 0).sort_values(by="Rating")
```

# Convert datatype in dataframe
<br>Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html?highlight=to_numeric#pandas.to_numeric

```python
data_phones["Rating"] = data_phones["Rating"].astype(str)
```

```python
data_phones["Rating"] = data_phones["Rating"].apply(pd.to_numeric, errors='ignore')
```

# Rolling calcs, pct change
<br> Roll documentation: https://pandas.pydata.org/docs/reference/api/pandas.core.window.rolling.Rolling.mean.html?highlight=roll
<br> Pct change documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html?highlight=pct#pandas.DataFrame.pct_change

```python
data_phones["Rolling rating"] = data_phones["Rating"].rolling(2).mean()
```

```python
data_phones["PCT change of rating"] = data_phones["Rating"].pct_change()
```
