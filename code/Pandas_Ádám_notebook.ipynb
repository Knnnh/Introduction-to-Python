# Merging dataframes
<br>Documentation is here: https://pandas.pydata.org/docs/user_guide/merging.html

```python
#We create new dataframes, which will include brand name and review columns for Samsung and Nokia
samsung_reviews = data_phones[data_phones["Brand Name"] == "Samsung"][["Brand Name","Reviews"]].reset_index(drop=True)
nokia_reviews = data_phones[data_phones["Brand Name"] == "Nokia"][["Brand Name","Reviews"]].reset_index(drop=True)
```

```python
#concat function is useful if you just want to concatenate dataframes along indexes
#if you use axis=0, it will concatenate vertically, i.e. add dataframe below each other
#if you use axis=1, it will concatenate horizontally, i.e. add dataframe next to each other
pd.concat([samsung_reviews,nokia_reviews], axis = 1)
```

```python
#We create a function that will create a dataframe with rating distribution, sorted from lowest to highest
#In the function, you have to add the brand name you want to do this operation for

def create_rate_dist_df(brand_name):
    #We check the occurrence of rating unique values for the specified brand in the function
    review_rate = data_phones[data_phones["Brand Name"] == brand_name]["Rating"].value_counts()
    
    #we reset_index to get a new column of unique values 
    review_rate = pd.DataFrame(review_rate).reset_index()
    
     #We add a new column that will add a brand name for each row. You can do it by multiplying the brand name string in the list
    review_rate["Type"] = [brand_name]*len(review_rate)
    
    #You can rename both columns using the rename function and using dictionary
    review_rate = review_rate.rename(columns = {"index":"Rating", "Rating":"Number of rating"})
    
    #You can sort values by rating. By default, it will show you from lower to higher
    review_rate = review_rate.sort_values(by="Rating")
    
    #We create a new column to calculate the % of each value in the dataframe. We just divide each value by the sum of the column.
    review_rate["% of total rates"] = review_rate["Number of rating"]/review_rate["Number of rating"].sum()
    
    #We reset the index to start the dataframe from 0 and increase by one
    review_rate = review_rate.reset_index(drop=True)
    
    #The function will return this dataframe
    return review_rate
```

```python
#In a list, we specify the brand names we want to run the function
types = ["Nokia", "Samsung"]

#We'll do a for loop and the output will be a dictionary. 
#The keys are the brand names from the list above and the values are the dataframes
dataframe_df = {}

#We run the for loop for the list and save it into the list
for i in types:
    dataframe_df[i] = create_rate_dist_df(i)
```

```python
#If you call one key from the dictionary, it'll show you the dataframe
dataframe_df["Nokia"]
```

```python
#We have two dataframes that have identical dataframe structure
#We use concat to merge them together and we sort the dataframe by Rating
pd.concat([dataframe_df["Nokia"],dataframe_df["Samsung"]], axis = 0).sort_values(by="Rating")
```

Below, I created a new dataframe to explain the difference between pd.concat and pd. merge. 
<br>We created three different dataframes, where the first column is about the country, and the order of the countries is different in each dataframe.
<br>The second column shows some basic economical data.

```python
#We can manually create dataframes by using dictionaries. 
#Keys are the column names, and the values should be a list, which are the rows
df_1 = pd.DataFrame({
    "Country": ["USA", "China", "UK", "Germany", "Japan"],
    "2021 GDP (USD bn)": [23000, 18000, 3200, 4200, 5000]})
df_1
```

```python
#In this dataframe, the order of the countries are different than in df_1
df_2 = pd.DataFrame({
    "Country": ["China","UK","USA","Germany"],
    "2021 Population (in mn)": [1400, 67, 331, 83]})
```

```python
df_3 = pd.DataFrame({
    "Country names": ["China","UK","USA","Germany"],
    "2022 Inflation YoY (%)": [2.1, 1.6, 0.4, 0.9]})
```

```python
#If you use pd.merge, it will pair your dataframe's values next to each other.
#It's very similar to a vlookup() function in excel
#As a parameters, you have to add the dataframes you want to merge. 
#"on" parameter will set the column by which the dataframe values are merged.
#"how" parameter will define how to merge. If you set intersection, it'll merge only values that can be found in both dataframes
#If you set outer, it'll merge all values. For example, Japan value is in df_1, but not in df_2, so there is NaN for population data
pd.merge(df_1,df_2,on="Country", how="outer") 
```

```python
#Another way to use merge if you first set the dataframe's parameter you want to merge with another dataframe
#As a parameter, you have to define the dataframe you want to merge with your second dataframe
#Similarly as above, "on" parameter will set the column by which the dataframe values are merged.
df_1.merge(df_2, on ="Country")
```

```python
#As you can see below, concat just concatenates dataframes along indexes and doesn't pair the similar values 
pd.concat([df_1,df_2],axis=1)
```

```python
#Join function is very similar to concat, it'll merge dataframes along indexes and doesn't pair the similar values
df_1.join(df_3,how="outer")
```

# Rolling calcs, pct change
<br> Roll documenttion: https://pandas.pydata.org/docs/reference/api/pandas.core.window.rolling.Rolling.mean.html?highlight=roll
<br> Pct change documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html?highlight=pct#pandas.DataFrame.pct_change

```python
#Rolling function is great if you want to run an operation on a rolling basis
#Below, we calculate the average using a rolling window of 2
#You can specify the window in the () of the rolling() function
data_phones["Rolling rating"] = data_phones["Rating"].rolling(2).mean()
```
